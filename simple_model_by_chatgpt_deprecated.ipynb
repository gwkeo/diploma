```python
import pandas as pd
import os

path = 'russian-news-telegram-dataset.csv'
if os.path.exists(path):
    df = pd.read_csv(path)
else:
    df = pd.read_json("hf://datasets/ScoutieAutoML/russian-news-telegram-dataset/scoutieRussianNewsTelegramDataset.json", lines=True)
    df.to_csv('russian-news-telegram-dataset.csv')
df
```

```python
# df_sampled = df[(df.create_time >= '2024-01-01') & (df.create_time < '2024-02-01')]
df_sampled = df
df_sampled
```

```python
df_sampled['create_time'] = pd.to_datetime(df_sampled['create_time'])
df_sampled = df_sampled.sort_values('create_time')
df_sampled
```

```python
ts = df_sampled.set_index('create_time').resample('1h').size()
ts = ts.fillna(0)
ts
```

```python
ts.describe()
```

```python
ts_smooth = ts.rolling(3, center=True).mean()
```

```python
median = ts_smooth.rolling(12).median()
mad = (ts_smooth-median).abs().rolling(12).median()

score = (ts_smooth - median) / (mad + 1e-6)
events_mask = score > 4
```

```python
events_mask
```

```python
import matplotlib.pyplot as plt

plt.plot(ts_smooth)
plt.scatter(ts_smooth.index[events_mask], ts_smooth[events_mask], color='red')
```

```python
peak = ts_smooth.idxmax()
window = ts_smooth[peak - pd.Timedelta('12H'): peak + pd.Timedelta('24H')]
window
```

```python
decay = window[window.index > peak]
decay = decay[decay > 0]
decay
```

```python
t = (decay.index - decay.index[0]).total_seconds() / 3600
y = decay.values
t, y
```

```python
import numpy as np

def exp_model(t, A, lam, C):
    return A * np.exp(-lam * t) + C

def power_model(t, A, alpha):
    return A * (t + 1) ** (-alpha)
```

```python
from scipy.optimize import curve_fit

params_exp, _ = curve_fit(exp_model, t, y, maxfev=10_000)
params_power, _ = curve_fit(power_model, t, y, maxfev=10_000)
```

```python
from sklearn.metrics import mean_squared_error

y_exp = exp_model(t, *params_exp)
y_power = power_model(t, *params_power)

rmse_exp = np.sqrt(mean_squared_error(y, y_exp))
rmse_power = np.sqrt(mean_squared_error(y, y_power))
```

```python
rmse_exp, rmse_power
```

```python
model = "exp" if rmse_exp < rmse_power else "power"
model
```

```python
plt.plot(t, y, label='real')
plt.plot(t, y_exp, label='exp')
plt.plot(t, y_power, label='power')
plt.legend()
```
